```python
import pandas as pd
import os
import json
```

```python
Q = []
A = []

lst = os.listdir()
full_dirs = [s for s in lst if '멀티모달' in s]

for part in full_dirs:
    dir_lst = os.listdir(part)
    for dir in dir_lst:
        path = part + '/' + dir
        filenames = os.listdir(path)
        
        find_word = '_interpolation.json'
        lst = [s for s in filenames if find_word in s]
        
        filename = min(lst, key=len)
        with open(path + '/' + filename, 'r', encoding='utf-8') as f:
            json_object = json.loads(f.read())
        
        st_time = {}
        for idx, dialog in enumerate(json_object['dialogue_infos']):
            st_time[idx] = dialog['start_time']
        sort_time = sorted(st_time.items(), key = lambda item: item[1])
        sort_time = [i[0] for i in sort_time]

        for i in range(len(sort_time)):
            try:
                now = sort_time[i]
                next = sort_time[i+1]
                Q.append(json_object['dialogue_infos'][now]['utterance'])
                A.append(json_object['dialogue_infos'][next]['utterance'])
            except:
                pass
```

```python
script = pd.DataFrame([Q, A]).T
script.columns = ['Q','A']
script
```

```python
script.to_csv('script_part_5.csv', encoding='utf-8', index=False)
```

#### Merge All Data

```python
import pandas as pd
import os
```

```python
path = 'data/Script'
filenames = os.listdir(path)
```

```python
data = pd.DataFrame()
for file in filenames:
    try:
        temp = pd.read_csv(path+'/'+file, encoding='utf-8')
    except:
        temp = pd.read_csv(path+'/'+file, encoding='cp949')
    data = pd.concat([data, temp], axis=0)
data
```

```python
data.to_csv('Script_for_all.csv', encoding='utf-8', index=False)
```
